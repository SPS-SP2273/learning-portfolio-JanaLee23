Exercise 3

|  # | Term         | Description |
|:--:|--------------|:-----------|
|  1 | Function     | Reusable block of code that does a specific task and can be called by name |
|  2 | Arguments    | Values passed into a function |
|  3 | Comments     | Ignored by python, summarises what the code does for humans who are assumed to have knowledge of coding |
|  4 | Indentations | Spaces at beginning of a line that defines code blocks |
|  5 | Packages     | Collection of modules (functions, variables, classes, constants) that provide functionality beyond in-built python |
|  6 | `.`          | Indicates ownership, i.e. accesses attributes that belong to an object or package |
|  7 | `:`          | Indicates start of an indented code block, e.g. after `if` or function definitions |
|  8 | `[]`         | Empty list |
|  9 | `[[],[]]`    | List containing two empty lists |
| 10 | `{}`         | Empty dictionary |
