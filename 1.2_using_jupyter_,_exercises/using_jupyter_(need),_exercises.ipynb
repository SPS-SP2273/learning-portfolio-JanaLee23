### Introduction

The general form of a quadratic equation is:
$$ax^2 + bx + c = 0$$

### Solutions


Provided $a \neq 0$, we can use an elementary algebraic method called <u>[completing the square](https://en.wikipedia.org/wiki/Completing_the_square)</u> to show that a quadratic equation has the following solution:
$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

### Discriminant

The quantity $\bigtriangleup = b^2-4ac$ is called the **discriminant** of the equation and decides the nature of its solutions. The table below shows the various possibilities.

| Discriminant    |  Roots   |
| :--: | :--- |
| $\bigtriangleup = b^2-4ac = 0$   |  A single solution of $-b/(2a)$  |
| $\bigtriangleup = b^2-4ac > 0$   |  Two distinct solutions  |
| $\bigtriangleup = b^2-4ac < 0$   |  No real solutions; both are complex.  |
