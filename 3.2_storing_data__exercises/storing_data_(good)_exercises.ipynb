# Exercise 2

```python
import numpy as np

np_array_2d = np.array([[1, "A"], [3, "C"], [2, "B"], [4, "D"],
                        [5, "E"], [7, "G"], [6, "F"], [8, "H"],
                        [10, "J"], [9, "I"]])

column_1 = np_array_2d[:, 0]
column_1_int = column_1.astype(int)

np_mask = column_1_int % 2 == 1
column_2_odd = np_array_2d[np_mask, 1]

print(column_2_odd)
```
