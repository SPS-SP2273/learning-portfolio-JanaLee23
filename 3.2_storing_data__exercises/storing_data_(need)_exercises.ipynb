# Exercise 5

$$r =  \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2} \sqrt{\sum (y_i - \bar{y})^2}}$$

```python
import numpy as np

x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]

# Convert x_1 and x_2 to numpy arrays for easier arithmetic operation
x_1_np = np.array(x_1)
x_2_np = np.array(x_2)

# means 

x_1_mean = np.mean(x_1_np)
x_2_mean = np.mean(x_2_np)
print(x_1_mean)
print(x_2_mean)

#std devs
x_1_std = np.std(x_1_np)
x_2_std = np.std(x_2_np)
print(x_1_std)
print(x_2_std)

# calculating Pearson correlation coefficient r manually

x_1_residual = x_1_np - x_1_mean
x_2_residual = x_2_np - x_2_mean

x_1_residual_squared_sum = np.sum(x_1_residual * x_1_residual)
x_2_residual_squared_sum = np.sum(x_2_residual * x_2_residual)

sum_residual_crossmultiplied = np.sum(x_1_residual * x_2_residual)

# Manual r
print(sum_residual_crossmultiplied / (x_1_residual_squared_sum * x_2_residual_squared_sum)**0.5)
# np.corrcoef() r
print(np.corrcoef(x_1_np, x_2_np))
```

Similarity:
The value of the r between x_1 and x_2 are the same for both the manual and actual calculation

Difference:
np.corrcoef() function in Python returns a numPy array that is a symmetric matrix of Pearson correlation coefficients, while the manual calculation returns r directly between x_1 and x_2
