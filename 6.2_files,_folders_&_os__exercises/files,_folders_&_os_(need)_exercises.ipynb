Task 1

```python
month_dictionary = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

# Use the dictionary to convert 'oct-08_singapore.txt' to 10-08-singapore.txt
old_file_name_example = 'oct-08_singapore.txt'

# Splits old_filename into its components
month, day_city = old_file_name_example.split('-')
day, city_extension = day_city.split('_')

new_file_name = month_dictionary[month] + '-' + day + '-' + city_extension
print(new_file_name)
```

Task 2

```python
def rename_my_file(old_file_name):
    month, day_city = old_file_name.split('-')
    day, city_extension = day_city.split('_')

    new_file_name = month_dictionary[month] + '-' + day + '-' + city_extension
    return new_file_name

# Eg
print(rename_my_file('oct-08_singapore.txt'))  
```

Task 4 

```python
import os
import glob

# Use glob to rename .txt files in os-collaboration-exercise-data directory

directory = 'os-collaboration-exercise-data'

# Self check of files in the directory
# print(glob.glob('os-collaboration-exercise-data/*.txt', recursive=True))

for old_file_path in glob.glob(os.path.join(directory, '*.txt')):
    try:
        # Extract basename of old_file_path to pass into the rename_my_file function
        old_file_name = os.path.basename(old_file_path)
        
        new_file_name = rename_my_file(old_file_name)
        new_file_path = os.path.join(directory, new_file_name)
        os.rename(old_file_path, new_file_path)

    except FileExistsError:
        print(f"{new_file_name} already exists; skipping creation.")
    
# Manually checked: Files have been renamed in the directory.
```

```python
import os

os.mkdir("cities")

cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    try:
        city_path = os.path.join("cities", city)
        os.mkdir(city_path)
        print(f"Created directory: {city_path}")
    except FileExistsError:
        print(f"{city_path} already exists; skipping creation.")
```

Task 6 

```python
import os
import glob
import shutil

# Use glob to list all the files from Cairo.

city_path_cairo = os.path.join("cities", "Cairo")  
city_files_cairo = glob.glob('os-collaboration-exercise-data/*cairo.*')


for file_path_cairo in city_files_cairo:
    shutil.copy(file_path_cairo, city_path_cairo)
```

Task 7

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    city_path = os.path.join("cities", city)  
    city_files = glob.glob(f'os-collaboration-exercise-data/*{city}.*')

    for file_path in city_files:
        dest_path = os.path.join(city_path, os.path.basename(file_path))
        if not os.path.exists(dest_path):
            shutil.move(file_path, city_path)
```
