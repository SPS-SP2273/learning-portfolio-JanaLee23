# 4 Never compare floats directly

## 4.1 The Problem

```python
a = 0.1
a3 = 0.3
a * 3 == a3

f'{0.3:.17f}'
```

## 4.2 A solution

```python
#solution 1
eps = 1E-10
abs(a * 3 - a3) < eps
```

```python
#solution 2
import numpy as np
np.isclose(a * 3, a3)
```

# 5 Combining English and variables

```python
name = "Batman"
print(f"Hello {name}!")
print(f"Hello {name.upper()}!")

x = 10
print(f"The value of {x} squared is {x**2}!")
```

```python
text = 'Bruce Wayne is Batman.'
print(f'{text}')
print(f'{text:>30}')      # A block of 30 characters;
                          # aligned right
print(f'{text:^30}')      # A block of 30 characters;
                          # aligned centre
print(f'{text:<30}')      # A block of 30 characters;
                          # aligned left

print(f'The cube of pi to 6 decimal places is {np.pi**3:.6f}')
print(f'The cube of pi to 6 decimal places is {np.pi**3:.6e}')
```

## 5.1 Structure of f-strings

f-string formatting has the structure {X:>0Y.ZW}

| Letter | Action                 | Possible Options |
|:------:|----------------------|-----------------|
| X      | Variable to format     | Can be a number or a string |
| >      | Alignment             | `<` Left justified<br>`>` Right justified<br>`^` Centre justified |
| 0      | Use 0â€™s to pad spaces | You can use other characters like a space |
| Y      | Total number of characters | |
| Z      | Number of decimal places | |
| W      | Specifies the type of variable | `f` (float)<br>`d` (integer)<br>`s` (string)<br>`g` (Python decides automatically) |

| Escape Sequence | Meaning        |
