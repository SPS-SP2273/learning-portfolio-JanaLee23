# Exercise 1

```python
%%timeit
square_odd = []
for i in range(10):
    if i % 2 == 0:
        continue
    square_odd.append(i**2) 
print(square_odd)
```

# Exercise 2

```python
%%timeit
[i**2 for i in range(10) if i % 2 != 0]
```

# Exercise 3

Growing list method
14.8 μs ± 2.19 μs per loop

List comprehension
941 ns ± 28.9 ns per loop

List comprehension is faster. One reason is that there are there is the python function call of .append() for the growing list method, while there isn't for the list comprehension method since the same function is done internally in C.

# Exercise 4

```python
quantity = 100
year = 0

while quantity > 50: # lazy to make a new variable for 50
    quantity *= 0.95
    year += 1

print(f"{year} years")
```

```python
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]

n = len(co2_data)

for i in range(n-1): #n-1 since there's n-1 differences in ppm
    difference = co2_data[i+1][1] - co2_data[i][1]
    if (difference >= 3):
        print(co2_data[i+1][0], difference)
```
