# Exercise 1

```python
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]

for c in temperatures_celsius:
    # converts each temperature to Fahrenheit
    f = c  * 9/5 + 32
    # prints both temperatures as a Celsius, Fahrenheit pair in tuple form
    print((c, f))
```

```python
for i in range(1, 6):
    print(f'{i} :', end='\t')
    for j in range(1, 6):
        print(f"{i*j}", end='\t')
    print()
```
