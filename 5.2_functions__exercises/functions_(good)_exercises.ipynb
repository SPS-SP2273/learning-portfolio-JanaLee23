# Exercise 1

```python
# Function converts temperatures from 
# Celsius to either Fahrenheit or Kelvin.

def convert_celsius(temperature_celsius, target_scale: str = 'Fahrenheit'):
    try: 
        if target_scale == 'Kelvin':
            return temperature_celsius + 273.15
        else:
            return temperature_celsius * 9/5 + 32
    except TypeError:
        return "Please enter correct numerical temperature_celsius."

# Eg 
print(convert_celsius(30, 'Kelvin'))
print(convert_celsius(30, 'Fahrenheit'))
print(convert_celsius(30)) # Default target_scale set to 'Fahrenheit'
print(convert_celsius('string', 'Kelvin'))
```

# Exercise 2

```python
def convert_fahrenheit(temperature_fahrenheit, target_scale: str = 'Celsius'):
    try:
        if target_scale == 'Kelvin':
            return (temperature_fahrenheit - 32) *5/9 + 273.15
        else:
            return (temperature_fahrenheit - 32) *5/9 
    except TypeError:
        return "Please enter correct numerical temperature_fahrenheit."

print(convert_fahrenheit(86, 'Kelvin'))
print(convert_fahrenheit(86, 'Celsius'))
print(convert_fahrenheit(86)) #  Default target_scale set to 'Celsius'
print(convert_fahrenheit('string', 'Kelvin'))
```

# Exercise 3

```python
def convert_kelvin(temperature_kelvin, target_scale: str = 'Celsius'):
    try:
        if target_scale == 'Fahrenheit':
            return (temperature_kelvin - 273.15) * 9/5 + 32
        else:
            return temperature_kelvin - 273.15    
    except TypeError:
        return "Please enter correct numerical temperature_kelvin."

def convert_temperature(temperature, source_scale: str, target_scale: str):
    if source_scale == target_scale:
        return temperature
    elif source_scale == 'Kelvin':
        return convert_kelvin(temperature, target_scale)
    elif source_scale == 'Celsius':
        return convert_celsius(temperature, target_scale)
    elif source_scale == 'Fahrenheit':
        return convert_fahrenheit(temperature, target_scale)
    else:
        return "Unsupported source scale."
    
# Eg
print(convert_temperature(303.15, 'Kelvin', 'Fahrenheit'))
```
