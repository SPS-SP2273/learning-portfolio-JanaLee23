# Exercise 1

```python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'
```

if name != 'Batman', then the function just returns f'Hello {name}!'. The second return works like an else because once a return runs, the function exits. (I.e. "When Python sees a return keyword it jumps out of the function with the return value.")

# Exercise 2

```python
def calculate_bmi(weight, height):
    BMI = weight / (height ** 2)
    if BMI >= 30:
        return "Obese"
    elif BMI >= 25:
        return "Overweight"
    elif BMI >= 18.5:
        return "Normal weight"
    else:
        return "Underweight"  

# Eg if user puts numerical values for bmi as intended 
# (lazy to do raise TypeError() and ValueError())    
print("BMI Category:", calculate_bmi(50, 1.6))
```

# Exercise 3

```python
def factorial(n):
    if n < 0:
        return "Invalid input"
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n+1):
            result *= i
        return result

# Eg
print(factorial(4))
```
